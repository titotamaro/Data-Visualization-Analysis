```python
df4 = pd.DataFrame({
    'key':['K0','K1','K2','K3'],
    'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3']
})
df4
```

```python
df5 = pd.DataFrame({
    'key':['K0','K1','K2','K3'],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3']
})
df5
```

```python
pd.merge(df4,df5,how='inner',on='key') # Seperti SQL Inner Join
```

```python
pd.merge(df4,df5,how='left',on='key')
```

```python
pd.merge(df4,df5,how='outer',on='key')
```

```python
pd.merge(df4,df5,how='right',on='key')
```
