```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.isna().sum() # menunjukkan berapa yg kosong
```

```python
tips.head() 
```

```python
tips.info() # menunjukkan berapa yang isi
```

```python
### distplot - Univariate (1 variabel)
### KDE = Kernel Density Estimation

sns.distplot(tips['total_bill'])
plt.show()
```

```python
sns.distplot(tips['total_bill'], kde=False)
plt.show()
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=20)
plt.show()
```

```python
## Bivariate (2 atau lebih variabel)

sns.jointplot(tips['total_bill'],tips['tip'], kind='scatter')
plt.show()
```
