```python
Date_Purchase = data['Date of Purchase'].dt
```

```python
data['Tahun'] = Date_Purchase.year
```

```python
data['Bulan'] = Date_Purchase.month
```

```python
data['Tanggal'] = Date_Purchase.day
```

```python
data['Hari'] = Date_Purchase.dayofweek # 0 itu senin, mulai dari0 sampai 6b
```

```python
data
```
