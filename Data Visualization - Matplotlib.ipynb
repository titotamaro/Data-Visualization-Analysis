## Cari titik potong dari 
3x + y = 15
x + y = 7

```python
A = np.array([[3,1],[1,1]])
B = np.array([15,7])
hasil = np.linalg.solve(A,B)
hasil
```

```python
# 3x + y = 15 --> if x=0, y = 15
           #  --> if y=0, x = 5

# x + y = 7   --> if x=0, y = 7
           #  --> if y=0, x = 7
    
```

```python
x = (0,5)
y = (15,0)
plt.plot(x,y)
```

```python
x2 = np.arange(0,7)
y2 = np.arange(7,0)
plt.plot(x2,y2)
```

```python
plt.plot(x,y,'-.',x2,y2,4,3,'yo',)
plt.title('Titik Potong')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.grid(True)
plt.legend(['3x+y=15','x+y=7'],loc=0)
   
# for i in x:
#     if(15-3*i)==(7-i):
#         plt.plot(i,(7-i),'ro')
#         plt.text(i+.1,(7-i)+.5)
        
for i in y:
    plt.text(x[list(y).index(i)]-0.2,i+3,i)
        
plt.annotate('Titik Potong! (4,3)',
             xy=(4,3), xytext=(5,5),
            arrowprops = {
                'facecolor':'red',
                'shrink':0.1,
                'width':5,
                'headwidth':20,
                'headlength':15
            })
plt.show()
```

```python
# Pie Chart --> Untuk komposisi
```

```python
persentase = [25,60,15]
nama = ['Andi','Rudi','Deni']

plt.pie(
    persentase, 
    labels=nama,
    colors=['r','g','y'],
    startangle= 0,
    shadow=True,
    textprops={
        'size':20 ,
        'color':'k'
    },
#     autopct='%i%%' # integer format
    autopct='%1.1f%%' # float format
)
plt.show()
```

```python
persentase = [25,60,15]
nama = ['Andi','Rudi','Deni']

plt.pie(
    persentase, 
    labels=nama,
    colors=['r','g','y'],
    startangle= 0,
    shadow=True,
    textprops={
        'size':10 ,
        'color':'k'
    },
#     autopct='%i%%' # integer format
    autopct='%1.1f%%', # float format
    explode=(.1,0.3,3)
)
plt.show()
```

```python
persentase = [25,60,15]
nama = ['Andi','Rudi','Deni']

plt.pie(
    persentase, 
    labels=nama,
    colors=['r','g','y'],
    startangle= 0,
    shadow=True,
    radius = 2,
    textprops={
        'size':20 ,
        'color':'k'
    },
#     autopct='%i%%' # integer format
    autopct='%1.1f%%', # float format
    explode=(.1,0.3,3),
)

plt.show()
```
